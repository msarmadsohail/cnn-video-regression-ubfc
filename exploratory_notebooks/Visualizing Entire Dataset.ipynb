## Visualizing the entire dataset

### Traces --- Row 1

```python
import numpy as np
import matplotlib.pyplot as plt
import os

dataset_path = '/content/drive/MyDrive/UBFC_DATASET/DATASET_2'
subject_dirs = [d for d in os.listdir(dataset_path) if os.path.isdir(os.path.join(dataset_path, d))]

for subject_dir in subject_dirs:
    file_path = os.path.join(dataset_path, subject_dir, 'ground_truth.txt')
    if os.path.exists(file_path):
        data = np.loadtxt(file_path)
        traces = data[0, :]

        plt.figure()
        plt.plot(traces)
        plt.title(subject_dir)
        plt.xlabel('Time')
        plt.ylabel('Heart Rate')

        min_trace = np.min(traces)
        max_trace = np.max(traces)
        length = len(traces)

        plt.text(0.95, 0.85, f'Min Trace: {min_trace:.2f}\nMax Trace: {max_trace:.2f}\nLength: {length}',
                 horizontalalignment='right', verticalalignment= 'top', transform=plt.gca().transAxes)

        # plt.savefig(os.path.join(dataset_path, subject_dir + '.png'))
        plt.show()
        plt.close()
```

### Heart Rates --- Row 2

```python
import numpy as np
import matplotlib.pyplot as plt
import os

dataset_path = '/content/drive/MyDrive/UBFC_DATASET/DATASET_2'
subject_dirs = [d for d in os.listdir(dataset_path) if os.path.isdir(os.path.join(dataset_path, d))]

for subject_dir in subject_dirs:
    file_path = os.path.join(dataset_path, subject_dir, 'ground_truth.txt')
    if os.path.exists(file_path):
        data = np.loadtxt(file_path)
        heart_rates = data[1, :]

        plt.figure()
        plt.plot(heart_rates)
        plt.title(subject_dir)
        plt.xlabel('Time')
        plt.ylabel('Heart Rate')

        min_hr = np.min(heart_rates)
        max_hr = np.max(heart_rates)
        length = len(heart_rates)


        plt.text(0.95, 0.9, f'Min HR: {min_hr:.2f}\nMax HR: {max_hr:.2f}\nLength: {length}',
                 horizontalalignment='right', verticalalignment='top', transform=plt.gca().transAxes)

        # plt.savefig(os.path.join(dataset_path, subject_dir + '.png'))
        plt.show()
        plt.close()
```

### Row 3
